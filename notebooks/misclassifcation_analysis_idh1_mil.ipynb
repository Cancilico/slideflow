# Analysis misclassifcation w.r.t to mil models

```python

def analyze_by_feature_bags(df):
    # Define feature bags
    feature_bags = [
        'bags_ctranspath',
        'bags_histossl',
        'bags_plip',
        'bags_resnet50_imagenet',
        'bags_retccl'
    ]
    # Create a results dictionary to store results for each feature bag
    results = {}
    # Iterate through each feature bag
    for bag in feature_bags:
        misclassified = df[[col for col in df.columns if bag in col]].ne(df['y_true'], axis=0).sum(axis=1)
        print(misclassified)
        results[bag] = {
            'misclassification_count': misclassified.sum(),
            'misclassification_rate': (misclassified > 0).mean()
        }
    # Convert results to DataFrame for better visualization
    results_df = pd.DataFrame(results).T
    results_df.index.name = 'Feature Bag'
    return results_df

def analyze_by_models(df):
    # Define models
    models = [
        'model_attention_mil',
        'model_clam_mb',
        'model_clam_sb'
    ]
    # Create a results dictionary to store results for each model
    results = {}
    # Iterate through each model
    for model in models:
        misclassified = df[[col for col in df.columns if model in col]].ne(df['y_true'], axis=0).sum(axis=1)
        results[model] = {
            'misclassification_count': misclassified.sum(),
            'misclassification_rate': (misclassified > 0).mean()
        }
    # Convert results to DataFrame for better visualization
    results_df = pd.DataFrame(results).T
    results_df.index.name = 'Model'
    return results_df



```
