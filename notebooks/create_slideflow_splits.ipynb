```python
patient_df = pd.read_csv(r'D:\github\CLAM\dataset_csv\pandas_tumor_subtyping_train_val_clean807.csv')
split_0_df = pd.read_csv(r'D:\github\slideflow\notebooks\splits_0.csv')

split_config = {
    'strategy': 'fixed',
    'patients': {},
    'tfrecords': {
        'training': [],
        'validation': [],
        'test': []
    }
}

# Populate the patient details and the specific split categories
for index, row in patient_df.iterrows():
    patient_id = row['patient']
    slide_id = row['slide']
    label = row['label']

    patient_entry = {
        'outcome_label': label,
        'slides': [slide_id]
    }
    split_config['patients'][patient_id] = patient_entry

    # Determine which split the slide belongs to
    if slide_id in split_0_df['train'].values:
        split_config['tfrecords']['training'].append(slide_id)
    elif slide_id in split_0_df['val'].values:
        split_config['tfrecords']['validation'].append(slide_id)
    elif slide_id in split_0_df['test'].values:
        split_config['tfrecords']['test'].append(slide_id)

# Displaying the configuration for confirmation before saving
print(split_config)
```

```python
# Save the created split configuration to a JSON file
with open(r'D:\github\slideflow\notebooks\split_config.json', 'w') as file:
    json.dump(split_config, file, indent=4)
```

# 3. Eval test set SLIDEFlow  --> CLAM_SB

```python
eval_pred_path = 'D:/github/slideflow/panda_project/test_prediction.csv'
eval_df = pd.read_csv(eval_pred_path, index_col = 0)
eval_df
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, classification_report, f1_score, roc_auc_score
from sklearn.preprocessing import label_binarize


# Converting prediction probabilities into class predictions
predicted_probabilities = eval_df.filter(regex='y_pred').values
y_pred = np.argmax(predicted_probabilities, axis=1)
y_true = eval_df['y_true'].values

# For ROC AUC calculation, we need binary format of y_true
n_classes = predicted_probabilities.shape[1]
y_true_binary = label_binarize(y_true, classes=range(n_classes))
```

```python
overall_accuracy = accuracy_score(y_true, y_pred)
print("Overall Accuracy:", overall_accuracy)


f1 = f1_score(y_true, y_pred, average='weighted')  # Weighted average F1 score
print("Overall F1 Score:", f1)
f1_class = f1_score(y_true, y_pred, average=None)  # Per-class F1 score
print("F1 Score per Class:", f1_class)
```

```python
clf_report = classification_report(y_true, y_pred)
print("Classification Report:\n", clf_report)

roc_auc = roc_auc_score(y_true_binary, predicted_probabilities, multi_class='ovr')
print("Overall ROC AUC Score:", roc_auc)
```

## 4. Create dict structure as expect for the splits function in slideflow

```python
import pandas as pd
import json

# Load both CSV files
main_df = pd.read_csv(r'D:\github\CLAM\dataset_csv\pandas_tumor_subtyping_train_val_clean807.csv')
splits_df = pd.read_csv(r'D:\github\slideflow\notebooks\splits_0.csv')

# Initialize the dictionary structure for JSON
split_dict = {
    'strategy': 'fixed',
    'patients': {},
    'tfrecords': {
        'validation': [],
        'training': [],
        'test': []
    }
}

# Correctly initialize the keys for training, validation, and testing
for category in ['train', 'val', 'test']:
    split_dict['tfrecords'][category] = []

# Process training, validation, and test slides
for column in ['train', 'val', 'test']:
    slide_list = splits_df[splits_df[column].notna()][column].tolist()
    filtered_df = main_df[main_df['slide'].isin(slide_list)]
    for index, row in filtered_df.iterrows():
        patient_id = f"patient_{index}"  # Or another appropriate identifier
        slide_id = row['slide']
        label = row['label']
        
        # Append to the correct category in the dictionary
        split_dict['tfrecords'][column].append({
            'slide': slide_id,
            'patient_id': patient_id,
            'label': label
        })

        # Ensure patient details are stored correctly
        if patient_id not in split_dict['patients']:
            split_dict['patients'][patient_id] = {
                'outcome_label': label,
                'slides': [slide_id]
            }
        else:
            if slide_id not in split_dict['patients'][patient_id]['slides']:
                split_dict['patients'][patient_id]['slides'].append(slide_id)

# Save the created split configuration to a JSON file
with open(r'D:\github\slideflow\notebooks\split_config.json', 'w') as file:
    json.dump(split_dict, file, indent=4)
```

```python

# Load the JSON file
with open(r'D:\github\slideflow\notebooks\split_config.json', 'r') as file:
    loaded_splits = json.load(file)

# Print keys of the dictionary to understand its structure
print(loaded_splits.keys())

# Optionally, print more about the contents to understand the nested structures
for key, value in loaded_splits.items():
    print(f"{key}: {type(value)}")  # This will tell you what type each value is
    if isinstance(value, dict):
        print(f"  {key} contains: {value.keys()}")
```
